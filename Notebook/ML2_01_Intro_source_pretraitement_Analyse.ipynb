On voit que les variables  trestbps, chol, thalach, oldpeak et ca présentent des valeurs aberrantes. Passons a la correction de ces valeurs.

L'IQR mesure l'écart entre le premier quartile (Q1) et le troisième quartile (Q3) d'une distribution. On utilise cet intervalle pour détecter les valeurs aberrantes.

Formule de l'IQR :  **𝐼𝑄𝑅=𝑄3−𝑄1**
        
Une valeur est considérée comme aberrante si elle se situe :

        * .En dessous de : **Q1−1.5×IQR**
            
        * .Au-dessus de : **Q3+1.5×IQR**

## **Correction des valeurs aberrantes**

```python
outliers_var=["chol" ,"thalach","oldpeak","trestbps","ca"]

def appure(new_data_cap1, colonnes):
    for col in colonnes:
        # Copie pour conserver les valeurs avant traitement
        data_before = new_data_cap1.copy()

        # Calcul des bornes interquartiles
        percentile25 = new_data_cap1[col].quantile(0.25)
        percentile75 = new_data_cap1[col].quantile(0.75)
        iqr = percentile75 - percentile25
        upper_limit = percentile75 + 1.5 * iqr
        lower_limit = percentile25 - 1.5 * iqr

        # Remplacement des valeurs aberrantes par les bornes
        new_data_cap1 = new_data_cap1.copy()
        new_data_cap1[col] = np.where(
            new_data_cap1[col] > upper_limit, upper_limit,
            np.where(new_data_cap1[col] < lower_limit, lower_limit, new_data_cap1[col])
        )

        # Tracé des graphiques avant/après traitement
        plt.figure(figsize=(16, 8))

        # Histogramme + KDE avant
        plt.subplot(2, 2, 1)
        sns.histplot(data_before[col], kde=True, color="skyblue")
        plt.title(f"Histogramme + KDE de {col} (avant traitement)")

        # Boxplot avant
        plt.subplot(2, 2, 2)
        sns.boxplot(x=data_before[col], color="skyblue")
        plt.title(f"Boxplot de {col} (avant traitement)")

        # Histogramme + KDE après
        plt.subplot(2, 2, 3)
        sns.histplot(new_data_cap1[col], kde=True, color="lightgreen")
        plt.title(f"Histogramme + KDE de {col} (après traitement)")

        # Boxplot après
        plt.subplot(2, 2, 4)
        sns.boxplot(x=new_data_cap1[col], color="lightgreen")
        plt.title(f"Boxplot de {col} (après traitement)")

        plt.tight_layout()
        plt.show()

    return new_data_cap1

heart_disease_df = appure(heart_imputed_df,outliers_var)
```

La matrice de corrélation nous permet de remarquer qu'il n'existe pratiquement pas de relation entre les variables quantitatives. Ainsi, toutes les variables quantitatives seront retenues.
Nous nous interesserons à présent aux variables qualitatives.

```python
import pandas as pd
from scipy.stats import chi2_contingency
from itertools import combinations

# Liste des variables qualitatives
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']

# Parcours de toutes les paires de variables qualitatives
for var1, var2 in combinations(categorical_features, 2):
    # Création du tableau de contingence
    table = pd.crosstab(heart_disease_df_cat[var1], heart_disease_df_cat[var2])

    # Calcul du test de chi-deux
    chi2, p, dof, expected = chi2_contingency(table)

    print(f"Relation entre {var1} et {var2}:")
    print(f"  Chi² = {chi2:.2f}, p-value = {p:.4f}")
    if p < 0.05:
        print("  ➤ Relation significative (p < 0.05)")
    else:
        print("  ➤ Pas de relation significative")
    print("-" * 50)
```
